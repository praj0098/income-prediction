## Q.2 
- Plot a Sine Wave with color code with "black" color and style as dashdot, also mark "x" and "y" label and give a suitable title to it

```python
x=np.linspace(0,10,100)
plt.plot(x,np.sin(x),color="black",linestyle="-.")
plt.xlabel("X",size=40,color="b")
plt.ylabel("Y",size=40,color="b")
plt.title("SINE WAVE GRAPH",size=30,color="m")
```

## Q.4

-Plot Stacked Bar Graph for the above data.
    - Use proper colors and x and y label for the graph
    - Use legend positioned at upper right corner
    - Also mark xticks with an angle of 90 Degree

```python
Cities=['Delhi','Mumbai','Chennai','Kolkata','Hyderabad']
Clothing=np.array([82,78,48,46,39])
Equipment=np.array([48,31,51,29,18])
Accessories=np.array([15,20,14,13,9])
plt.bar(Cities,Clothing,color="black",bottom=Equipment+Accessories,label="Clothing")
plt.bar(Cities,Equipment,color="green",bottom=Accessories,label="Equipment")
plt.bar(Cities,Accessories,color="yellow",label="Accesories")
plt.xlabel("Stores",size=30)
plt.ylabel("Stocks",size=30)
plt.title("Store Vs Stocks",size=30)
plt.legend(loc="upper right")
plt.xticks(rotation=90)
```

## Q.10 For Tips Dataset
- Load the dataset using seaborn
- Draw the Strip Plot between 'total bill' and 'day'
- Rotate xticks with an angle of 45 Degree

```python
np.random.seed(50)
sns.stripplot(x='day',y='total_bill',data=df1,palette='Set1',alpha=0.8)
plt.title("Day vs Total Bill",size=30,color="m")
plt.xlabel('Day',size=30)
plt.ylabel('Total Bill',size=20)
plt.xticks(rotation=0)
```
